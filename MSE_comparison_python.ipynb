### MSE comparison

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define parameters
theta_mle = np.arange(0.1, 1.01, 0.01)
n = 1

# Calculate MSE for MLE
mse_mle = theta_mle * (1 - theta_mle) / n

# Calculate variance and bias for Bayes
var_bayes = (n * theta_mle * (1 - theta_mle)) / (n + 2)**2
bias_bayes = (n * theta_mle + n - theta_mle) / (n + 2)
mse_bayes = var_bayes + bias_bayes**2

# Plotting
plt.plot(theta_mle, mse_mle, label='MSE MLE', color='darkred')
plt.plot(theta_mle, mse_bayes, label='MSE Bayes', color='black')
plt.title('Comparison of MSE across all values of theta MLE and theta Bayes')
plt.xlabel('theta')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
